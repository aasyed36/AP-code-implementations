```python
import sys
from pathlib import Path

repo_root = Path("..").resolve()
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

import numpy as np
from sklearn.metrics import adjusted_rand_score

from apclust import (
    load_study_panels,
    StudyMatrices,
    run_ap_grid,
)
```

```python
data_dir = repo_root / "Data"

mainz_panels = load_study_panels(data_dir / "mainz_dict.npy")
trans_panels = load_study_panels(data_dir / "transbig_dict.npy")
vdx_panels = load_study_panels(data_dir / "vdx_dict.npy")

hk_studies = StudyMatrices(
    mainz=mainz_panels["HK_3"],
    transbig=trans_panels["HK_3"],
    vdx=vdx_panels["HK_3"],
)
```

```python
from apclust import run_multiple_ap, summarize_ap_runs
```

```python
# Mainz
runs = run_multiple_ap(
    hk_studies.mainz,
    n_runs=3,
    preference=None,
    damping=0.5,
    random_state=None,
)
summary = summarize_ap_runs(runs)
summary
```

```python
runs = run_multiple_ap(
    hk_studies.transbig,
    n_runs=3,
    preference=None,
    damping=0.5,
    random_state=None,
)
summary = summarize_ap_runs(runs)
summary
```

```python
runs = run_multiple_ap(
    hk_studies.vdx,
    n_runs=3,
    preference=None,
    damping=0.5,
    random_state=None,
)
summary = summarize_ap_runs(runs)
summary
```
